# Nettoyage du dataset 

### Importation du dataset

```python
import pandas as pd
import numpy as np


%time df = pd.read_csv('footscrap/data.csv')
df = df.dropna(thresh=4)

# Replace element

df = df.replace(['[]','-'], np.NAN)

#Drop Nan

df = df.dropna(subset=['price'])


def delextrachar(str):
    
        """
        Delete extra char
        
        """
        delr = str.replace("\r", "")
        delt = delr.replace("\t", "")
        delv = delt.replace(",", "")
        delvspace = delt.replace(" ", "")
        delvcrcg = delvspace.replace("[", "")   
        delvcrcd = delvcrcg.replace(" ", "")
        delvgui = delvcrcg.replace("'", "")
        final = delvgui.replace("]", "")
        
        return final
    
def virguleToPoint(word):
    delr = word.replace(",",".")
    return delr


def convert_string_to_list(srt):
    """
    Prends en entrée une string 
    Exemple : Ligue1, Coupe de la Ligue, Coupe de France
    Puis la convertit en liste 
    Exemple : [Ligue1, Coupe de la Ligue, Coupe de France]
    -----------
    Return : une liste de liste
    """
    for i, element in enumerate(srt):
        try:
            srt[i] = element.split(',')
        except:
            pass
    
    return srt   
```

### Nettoyage de la colonne poste 

```python
#Clean poste

liste_poste = []

for i, (poste1, poste2) in enumerate(zip(df['poste1'].values, df['poste2'].values)):
    
        if len(str(poste2)) < 5:
            liste_poste.append(poste1)
        else:
            liste_poste.append(poste2)

            
liste_poste  
df_poste = pd.DataFrame(liste_poste)
df_poste.columns = ['poste']

df_poste = df_poste.replace("[]",np.NaN)
df = df.drop(['poste1','poste2'],axis=1)
df = df.reset_index()
df = pd.concat([df, df_poste],axis=1)
   

df = df.dropna(subset=['poste','stats','competitions','age'])
```

```python
colrangeprice = [item.replace('K €', '1000') for item in df['price_range']]
colrangeprice = [item.replace('mio. €', '1000000') for item in colrangeprice]
colrangeprice = [float(i) for i in colrangeprice]

df['price'] = df['price'].apply(lambda x: delextrachar(x))
df['price'] = df['price'].apply(lambda x: virguleToPoint(x))

colprice = [float(i) for i in df['price'].values]

price = np.multiply(colrangeprice, colprice)
df_price = pd.DataFrame(price)
df_price.columns = ['price']
df = df.reset_index(drop=True)
df = df.drop(['price', 'price_range'], axis=1)
df = pd.concat([df, df_price], axis=1)
```

### Nettoyage de l'age

```python
def get_age(age):

    liste = []
    for i, element in enumerate(age):
        try :
            liste.append(int(element[element.find("(")+1:element.find(")")]))
        except:
            pass
    return liste

age_clean = get_age(df['age'])
dfage = pd.DataFrame(age_clean)
dfage.columns = ['age']
df = df.dropna(subset=['age'])
df = df.reset_index(drop=True)
df['age'] = dfage['age']
```

### Nettoyage fin de contrat 

```python
def clean_fin_contrat(data):
    
    current_date = 2018
    result = []
    for i, element in enumerate(data):
        try:
            result.append(int(element.split('.')[2]) - current_date)
        except:
            result.append(np.NaN)
        
    return result

dfcontrat = pd.DataFrame(clean_fin_contrat(df['fin_contrat'].dropna()))
dfcontrat.columns = ['fin_contrat']
df = df.dropna(subset=['fin_contrat'])
df = df.reset_index(drop=True)

df = df.drop(['fin_contrat'],axis=1)
df = pd.concat([df, dfcontrat],axis=1)
```

### Nettoyage des stats

```python
df['poste'] = df['poste'].apply(lambda x: delextrachar(x))
df_values = df.values

#Values

stats = df['stats']
stats = stats.values

competitions = df['competitions']
competitions = competitions.values

competitions = convert_string_to_list(competitions)
stats = convert_string_to_list(stats)
```

```python
def generate_stats(stat, poste):
    """
    Convertit une liste de (4*Nombre de competition jouées + 1) en un tableau de données de
    dimension (Nombre de competition jouées + 1) X 4
    ------------------
    Return : une liste de stat par compétition
    """
    liste_of_stats = []
    
    if poste != 'Gardiendebut':
    
        for j in range(5,len(stat),4):
            liste_of_stats.append(stat[j:j+4])
        return liste_of_stats
    
    else:
        for j in range(4,len(stat),3):
            liste_of_stats.append(stat[j:j+3])
        return liste_of_stats
    
```

```python
def clean_stat(listofstats):

    """
    Convertit en float les données + remplace les tirets 
    par des 0

    """
    for i, stat in enumerate(listofstats):
        stat = [item.replace('-', '0') for item in stat]
        try:
            listofstats[i] = list(map(float, stat))
        except:
            pass
        
    return listofstats

stats_clean = clean_stat(stats) 
```

```python
def build_df_stats(stat_compet, compet,poste ):
    
    stcompetReg = [0,0,0,0]
    stcupReg = [0,0,0,0]
    stnatReg = [0,0,0,0]
    
    stcompetGK = [0,0]
    stcupReGK = [0,0]
    stnatReGK = [0,0] 
    
    mPlaycompet = 0
    mPlaycup = 0
    mPlaynat = 0
    
    if poste != 'Gardiendebut':
        
        for i, element in enumerate(compet):
          
            if element in championnat:
                stcompetReg = np.add(stcompetReg,stat_compet[i])
                
            elif element in coupe:
                
                stcupReg = np.add(stcupReg , stat_compet[i])
                
            elif element in continental:
                
                stnatReg = np.add(stnatReg, stat_compet[i])
                
        array_return = list(stcompetReg) + ['0','0']  + list(stcupReg) + ['0','0'] +  list(stnatReg) + ['0','0']
        
    else:
        
        for i, element in enumerate(compet):
            st = [item.replace('-', '0') for item in stat_compet[i]]
            if element in championnat:
                mPlaycompet += float(stat_compet[i][2]) 
                stcompetGK = np.add(stcompetGK, [float(i) for i in st[0:-1]])
               
            elif element in coupe:
                
                mPlaycup += float(stat_compet[i][2])
                stcupReGK = np.add(stcupReGK, [float(i) for i in st[0:-1]])
                
            elif element in continental:
                
                mPlaynat += float(stat_compet[i][2])
                stnatReGK = np.add(stnatReGK, [float(i) for i in st[0:-1]])
        
        array_return = ['0','0','0'] + [mPlaycompet] + [stcompetGK[0]] + [stcompetGK[1]] + ['0','0','0'] + [mPlaycup] + [stcupReGK[0]] + [stcupReGK[1]] + ['0','0','0'] + [mPlaynat] + [stnatReGK[0]] + [stnatReGK[1]]
        
    
        
    return array_return

matrix_of_stats = []

for i, (st, compet, poste) in enumerate (zip(stats_clean, competitions, df['poste'])):
    
    matrix_of_stats.append(build_df_stats(generate_stats(st, poste), compet, poste))
```
